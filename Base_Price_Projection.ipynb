```python
# Menghitung Q1 (kuartil pertama) dan Q3 (kuartil ketiga)
Q1 = fix['property_bedrooms'].quantile(0.25)
Q3 = fix['property_bedrooms'].quantile(0.75)

# Menghitung IQR (Interquartile Range)
IQR = Q3 - Q1

# Menentukan batas bawah dan batas atas untuk outlier
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Menghapus observasi yang mengandung outlier
fix_no_outliers = fix[(fix['property_bedrooms'] >= lower_bound) & (fix['property_bedrooms'] <= upper_bound)]

# Menampilkan data setelah menghapus outlier
print("\nData setelah menghapus outlier:")
print(fix_no_outliers)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np



# Data Anda
X = fix[['guests_max', 'min_stay', 'extra', 'property_type', 'property_bedrooms', 'area_name', 'booking_guest_number', 'bulan']]
y = fix['minimum_rate']

# Normalisasi data
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(np.array(y).reshape(-1, 1))

# Bagi data menjadi set pelatihan dan set pengujian
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)

# Inisialisasi model Random Forest
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Latih model
rf_model.fit(X_train, y_train.ravel())

# Prediksi harga untuk set pengujian
y_pred_scaled = rf_model.predict(X_test)

# Kembalikan nilai prediksi ke skala asli
y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1))
y_test_actual = scaler_y.inverse_transform(y_test)

# Evaluasi model
mse = mean_squared_error(y_test_actual, y_pred)
r_squared = r2_score(y_test_actual, y_pred)

print("Mean Squared Error:", mse)
print("R-squared (R^2):", r_squared)
rmse = np.sqrt(mse)
rmse
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Data Anda
X = fix[['minimum_rate', 'guests_max', 'min_stay', 'extra', 'property_type', 'property_bedrooms', 'area_name', 'booking_guest_number', 'bulan']]
y = fix['harga']

# Normalisasi data
scaler_X = MinMaxScaler()
scaler_y = MinMaxScaler()

X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(np.array(y).reshape(-1, 1))

# Bagi data menjadi set pelatihan dan set pengujian
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, random_state=42)

# Inisialisasi model Random Forest
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Latih model
rf_model.fit(X_train, y_train.ravel())

# Prediksi harga untuk set pengujian
y_pred_scaled = rf_model.predict(X_test)

# Kembalikan nilai prediksi ke skala asli
y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1))
y_test_actual = scaler_y.inverse_transform(y_test)

# Evaluasi model
mse = mean_squared_error(y_test, y_pred_scaled)
r_squared = r2_score(y_test, y_pred_scaled)

print("Mean Squared Error:", mse)
print("R-squared (R^2):", r_squared)

rmse = np.sqrt(mse)
print(rmse)
```

```python
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
import numpy as np
data=fix
# Ubah data menjadi format yang sesuai untuk LSTM
X = data[['minimum_rate', 'guests_max', 'min_stay', 'extra', 'property_type', 'property_bedrooms', 'area_name', 'booking_guest_number', 'bulan']].values
y = data['harga'].values

# Normalisasi data
scaler = MinMaxScaler(feature_range=(0, 1))
X_scaled = scaler.fit_transform(X)
y_scaled = scaler.fit_transform(y.reshape(-1, 1))

# Pisahkan data menjadi set pelatihan dan set pengujian
X_train, X_test = X_scaled[:int(0.8*len(X_scaled))], X_scaled[int(0.8*len(X_scaled)):]
y_train, y_test = y_scaled[:int(0.8*len(y_scaled))], y_scaled[int(0.8*len(y_scaled)):]

# Ubah data menjadi tiga dimensi untuk input model LSTM [jumlah sampel, jumlah time steps, jumlah fitur]
X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))

# Bangun model LSTM
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(LSTM(units=50))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')

# Latih model
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_test, y_test))

# Prediksi harga untuk set pengujian
y_pred_scaled = model.predict(X_test)

# Invers transformasi untuk mendapatkan harga aktual
y_pred = scaler.inverse_transform(y_pred_scaled)

# Evaluasi model (opsional)
mse = mean_squared_error(y_test, y_pred_scaled)
print("Mean Squared Error (MSE):", mse)
```
